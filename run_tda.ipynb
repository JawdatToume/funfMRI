```python
from fMRITDA import convertTDA, TopologicalfMRI, joinTDA
from labeler import labelsToFeatures
```

```python
# make topological data
convertTDA(fileType="perceptionTraining")
```

```python
# join tda objects for perception training data of subject 1
jTDA = joinTDA([1], fileType="perceptionTraining", save=True, str="VC_CORR")
```

```python
# alternatively load a pre-made one
import pickle
with open("trainedModels/subject_1_perceptionTraining_SHIFT_joinedPI.pickle", "rb") as f:
    jTDA = pickle.load(f)
```

```python
# get data from the object
data, labels = jTDA.getData()
```

```python
# expand features into full feature list given label and data set
features_exp = labelsToFeatures(labels, layer=22, labelType="train")
```

```python
# open a pre-trained model
with open("trainedModels/subject_1_Layer_22_Regresser_ridge.pickle", "rb") as f:
    modelTrained = pickle.load(f)
```

```python
# predict features, this may take a while
prediction = modelTrained.predict(data)
```

```python
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import ttest_1samp
res_vec = []
for i in range(features_exp.shape[1]):
    res_vec.append(pearsonr(features_exp[:,i], prediction[i,:]).statistic)
    
# conduct and print t-test, where the mean is zero
print(ttest_1samp(res_vec, 0, alternative="greater", nan_policy="omit"))
```
