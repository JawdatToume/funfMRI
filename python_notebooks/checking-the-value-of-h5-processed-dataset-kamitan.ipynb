```python
import os
import numpy as np
import h5py

def read_h5_file(file_path):
    try:
        # Open the HDF5 file in read-only mode
        with h5py.File(file_path, 'r') as file:
            # Access the dataset by name
            print(file.keys())
            for dataset_name in file.keys():
                dataset = file[dataset_name]
                
                print(dataset)
                try:
                    
                    dataset_values=dataset[()]
                    print(f"######VALUE STORED IN {dataset_name}#########")
                    print(dataset_values,"\n\n\n\n")
                except:
                    print(f"###############VALUE STORED IN {dataset_name}:###########")
                    
                    for value in dataset.values():
                        print(f"###############Values in metadata: {value}############")
                        for sub_values in value:
                            
                            print("\t\t\t",sub_values)
                                
                       
                    
            

            return 
    except Exception as e:
        print(f"Error reading HDF5 file: {e}")
        return None
```

```python
# Example usage:
file_path = '/kaggle/input/kamitani-fmri-alex-h5-features-mini-processed/ImageFeatures.h5'  # Replace with the path to your HDF5 file


# Read data from HDF5 file
data = read_h5_file(file_path)
```

```python

from bdpy import BData
file_path = '/kaggle/input/throwaway-data-from-ml-brain/ImageFeatures_pytorch.h5' 
# Create an empty BData instance
bdata = BData()

# Load BData from a file
bdata = BData(file_path)
```

```python

bdata.load(file_path)
```

```python
# Show 'key' and 'description' of metadata
bdata.show_metadata()
```

```python
# Get an array of voxel data in V1
data_cnn1 = bdata.select('layer1')  # shape=(M, num voxels in V1)

# `select` accepts some operators

data_cnn2 = bdata.select('layer2')
data_totaled = bdata.select('layer1 + layer2')

# Wildcard
data_visual = bdata.select('layer*')

# Get labels ('image_index') in the dataset
label_a  = bdata.select('*')
```
